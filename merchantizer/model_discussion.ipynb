### Notable Issues/Key Concerns
* Again, we see the doordashxcashapp problem rearing its ugly head. We had already addressed this with a TODO in the clean.py file, so that could be sorted in the future. I would recommend an investigation on multi-merchant descriptions.
* I would also consider looking into stopgap words that exist in merchant names. 365 Market is a good example, the results would likely be better if Market was removed from the 365 descriptions, thereby creating seperation between it and all the other market descriptions.
* My biggest concern was how computationally costly the bertopic package was. I had to downsample to 100 thousand from 10 million. About halfway through my task, I spun up a large GPU in the cloud and began running the same code with 1 million records. It is still running (~3 hours in). I would recommend an investigation into various strategies that could be leveraged (GPU acceleration, for example), to improve the compute time.
